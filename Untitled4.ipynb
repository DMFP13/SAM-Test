```python
import torch
from segment_anything import sam_model_registry, SamPredictor

# Load SAM ViT-B model
device = "cuda" if torch.cuda.is_available() else "cpu"
sam = sam_model_registry["vit_b"](checkpoint="weights/sam_vit_b_01ec64.pth")
sam.to(device)
predictor = SamPredictor(sam)
```

```python
from google.colab import files
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Upload an image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Get the uploaded image path

# Open and display the image
image = Image.open(image_path).convert("RGB")
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
# Convert to NumPy array for predictor
image_array = np.array(image)
predictor.set_image(image_array)
```

```python
# Example point prompt: (x, y) coordinate within the object you want to segment
input_point = np.array([[1000, 2000]])  # Example point coordinates
input_label = np.array([1])  # Label for the point (1 for foreground)

# Generate mask for the input point
masks, _, _ = predictor.predict(
    point_coords=input_point,
    point_labels=input_label,
    multimask_output=True  # Set to True if you want multiple mask proposals
)

# Display the generated mask
plt.imshow(masks[0], cmap="gray")
plt.axis("off")
plt.show()
```

```python
# Overlay the mask on the original image
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.imshow(image)
ax.imshow(masks[0], cmap="jet", alpha=0.5)  # Adjust alpha for mask transparency
plt.axis("off")
plt.show()
```
